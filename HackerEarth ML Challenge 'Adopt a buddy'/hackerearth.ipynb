# creating x and target variable

```python
x_train=data.drop(['pet_id','breed_category','pet_category'],axis=1)
y_breed_train=data['breed_category']
y_pet_train=data['pet_category']
```

```python
x_train.dtypes
```

# Encoding the 'color_type' column

```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,LabelEncoder
enc =OrdinalEncoder()
enc.fit(x_train[['color_type']])
a=enc.transform(x_train[['color_type']])
a
```

```python
x_train[['color_type']]=a
```

```python
print(x_train.shape)
print(y_breed_train.shape)
x_train=x_train.drop(['issue_date','listing_date'],axis=1)
```

```python
x_train
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
imp = IterativeImputer(max_iter=100, random_state=0)
#imp=KNNImputer(n_neighbors=30)
imp.fit(x_train)
x_train=imp.transform(x_train)
```

```python
print(x_train.shape)
x_train
```

# Applying the KNeighborsClassifier

```python
from sklearn.neighbors import RadiusNeighborsClassifier,KNeighborsClassifier
clf_breed = KNeighborsClassifier(3)
clf_pet= KNeighborsClassifier(3)
clf_breed.fit(x_train,y_breed_train)
clf_pet.fit(x_train,y_pet_train)
```

```python
print('breed score=',clf_breed.score(x_train,y_breed_train))
print('pet score=',clf_pet.score(x_train,y_pet_train))
```

# applying dicision tree classifier 

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf_breed=DecisionTreeClassifier(random_state=5, max_depth=30)
dt_clf_pet=DecisionTreeClassifier(random_state=5, max_depth=30)
dt_clf_breed.fit(x_train,y_breed_train)
dt_clf_pet.fit(x_train,y_pet_train)
```

```python
print('breed score=',dt_clf_breed.score(x_train,y_breed_train))
print('pet score=',dt_clf_pet.score(x_train,y_pet_train))
```

# applying neural network clasifier 

```python
from sklearn.neural_network import MLPClassifier
nn_clf_breed = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100), random_state=1,alpha=1, max_iter=250)
nn_clf_pet = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100), random_state=1,alpha=1, max_iter=250)
nn_clf_breed.fit(x_train,y_breed_train)
nn_clf_pet.fit(x_train,y_pet_train)
```

```python
print('breed score=',nn_clf_breed.score(x_train,y_breed_train))
print('pet score=',nn_clf_pet.score(x_train,y_pet_train))
```

# applying Gaussian naive_bayes classifier

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB
gnb_clf_breed=BernoulliNB(1,0)
gnb_clf_pet=BernoulliNB(1,0)
gnb_clf_breed.fit(x_train,y_breed_train)
gnb_clf_pet.fit(x_train,y_pet_train)
```

```python
print('breed score=',gnb_clf_breed.score(x_train,y_breed_train))
print('pet score=',gnb_clf_pet.score(x_train,y_pet_train))
```

# applying svm classifier

```python
from sklearn.svm import SVC
svm_clf_breed=SVC(kernel='rbf')
svm_clf_pet=SVC()
svm_clf_breed.fit(x_train,y_breed_train)
svm_clf_pet.fit(x_train,y_pet_train)
```

```python
print('breed score=',svm_clf_breed.score(x_train,y_breed_train))
print('pet score=',svm_clf_pet.score(x_train,y_pet_train))
```

```python
from sklearn.linear_model import SGDClassifier
sgdc_clf_breed = SGDClassifier(loss="hinge", penalty="l2", max_iter=10)
sgdc_clf_pet = SGDClassifier(loss="hinge", penalty="l2", max_iter=10)
sgdc_clf_breed.fit(x_train,y_breed_train)
sgdc_clf_pet.fit(x_train,y_pet_train)
```

```python
print('breed score=',sgdc_clf_breed.score(x_train,y_breed_train))
print('pet score=',sgdc_clf_pet.score(x_train,y_pet_train))
```
